```python
from sklearn.model_selection import train_test_split
```

```python
X = df_data.drop('Target', axis=1)
y = df_data['Target']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=5)
```

```python
from sklearn.svm import SVC
```

```python
svc = SVC()
```

```python
svc.fit(X_train,y_train)
```

```python
predictions = svc.predict(X_test)
```

```python
from sklearn.metrics import mean_squared_error,mean_absolute_error,r2_score, classification_report,confusion_matrix
```

```python
print(classification_report(y_test,predictions))
```

```python
cm = confusion_matrix(y_test, predictions)

sns.heatmap(cm,annot=True,fmt="d")
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X_train_scaled = scaler.fit_transform(X_train)
```

```python
X_test_scaled = scaler.fit_transform(X_test)
```

```python
model_scaled = svc.fit(X_train_scaled,y_train)
```

```python
predictions_scaled = model_scaled.predict(X_test_scaled)
```

```python
cm_scaled = confusion_matrix(y_test, predictions_scaled)

sns.heatmap(cm_scaled,annot=True,fmt="d")
```

```python
print(classification_report(y_test,predictions_scaled))
```

```python
param_grid = {'C': [0.1, 1, 10, 100], 'gamma': [1, 0.1, 0.01, 0.001], 'kernel': ['rbf']} 
```

```python
from sklearn.model_selection import GridSearchCV
```

```python
grid = GridSearchCV(SVC(),param_grid,refit=True,verbose=4)
```

```python
grid.fit(X_train_scaled,y_train)
```

```python
grid.best_params_
```

```python
grid.best_estimator_
```

```python
grid_predictions = grid.predict(X_test_scaled)
```

```python
cm_gamma = confusion_matrix(y_test, grid_predictions)
sns.heatmap(cm_gamma,annot=True,fmt="d")
```

```python
print(classification_report(y_test,grid_predictions))
```
